```python
baby_name_file = open('/Users/suman.nayak/MyProjects/PyML/Python/babynames/baby1990.html')
text = baby_name_file.read()    
```

```python
year_match = re.search(r'Popularity\sin\s(\d+)',text)
rank_tuple = re.findall(r'<tr align="right"><td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>',text)
```

```python
baby_names_list = []
```

```python
if not year_match:
    sys.stderr.write('Couldn\'t find the year!\n')
    sys.exit(1)
```

```python
baby_names_list.append(year_match.group(1))
```

```python
baby_names_list
```

```python
for each_tuple in rank_tuple:
    baby_names_list.append(each_tuple[1]+' '+each_tuple[0])
    baby_names_list.append(each_tuple[2]+' '+each_tuple[0])
```
